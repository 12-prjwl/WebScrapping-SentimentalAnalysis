```python
sia = SentimentIntensityAnalyzer()
sia
```

```python
# The original code to print all three sentiments at once


articles = []
pos = [] #1
neg = [] #2
neu = []

compound = [] #3
avgLen = [] #4
percentage_list = [] #5
avgWords = [] #6
fog = [] #7
complex_wc = [] #8
prsnl_pronoun = [] #9
wc = [] #10
avg_wLen = [] #11
for i in range(0,115):
   
    
    filename = f"{i+37}.txt" 
    with open(filename, 'r',encoding='utf-8',errors='replace') as f:
        toAnalyze = f.read()
        score = sia.polarity_scores(toAnalyze)

        articles.append(toAnalyze)
        



        neg.append(score['neg'])
        pos.append(score['pos'])
        neu.append(score['neu'])
        compound.append(score['compound'])
        
        
#-------------------------3(avg length)

    sentences = nltk.sent_tokenize(toAnalyze)

    
    sentence_lengths = [len(nltk.word_tokenize(sentence)) for sentence in sentences]

    
    average_length = sum(sentence_lengths) / len(sentence_lengths)
    avgLen.append(average_length)
     #-------------------------------------------- 4    (percentage)
    words = nltk.word_tokenize(toAnalyze)
    
    
    def count_syllables(word):
        return syllables.estimate(word)

    complex_words = [word for word in words if count_syllables(word) > 2]
    

    total_words = len(words)
    
  
    percentage = (len(complex_words) / total_words) * 100

    percentage_list.append(percentage)

    
    
#--------------------------------6
    average_words_per_sentence = total_words / len(sentences)
    
    avgWords.append(average_words_per_sentence)
#----------------5

    fog_index = 0.4 * (average_words_per_sentence + percentage)
    
    
    fog.append(fog_index)
    
    
#-------------------------------------------------------------------------
    
    
    complex_word_count = sum(1 for word in words if len(word) > 7)


    complex_wc.append(complex_word_count)

    
#----------------------------------------
    total_word_count = len(words)


    wc.append(total_word_count)
    
    
#---------------------------------------------------------------




            
    personal_pronouns = ['I', 'me', 'my', 'mine', 'myself',
                         'you', 'your', 'yours', 'yourself',
                         'he', 'him', 'his', 'himself',
                         'she', 'her', 'hers', 'herself',
                         'it', 'its', 'itself',
                         'we', 'us', 'our', 'ours', 'ourselves',
                         'they', 'them', 'their', 'theirs', 'themselves']

    # Calculate the count of personal pronouns
    personal_pronoun_count = sum(1 for word in words if word.lower() in personal_pronouns)
    
    prsnl_pronoun.append(personal_pronoun_count)
    
    
    #-----------------------------------------------------
    
    total_length = sum(len(word) for word in words)
    total_word_count = len(words)

        # Calculate the average word length
    average_word_length = total_length / total_word_count

    
    
    avg_wLen.append(average_word_length)

```

```python
# Calculate the subjectivity score for each compound score
subjectivity = []
for score in compound:
    subjectivity_score = abs(score)
    subjectivity.append(subjectivity_score)

print("Subjectivity Scores:", subjectivity)
```
